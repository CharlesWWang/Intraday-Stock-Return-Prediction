```python
# popular ones

from TwitterAPI import TwitterAPI, TwitterPager

SEARCH_TERM = '@Starbucks lang:en'

api = TwitterAPI('3Ya8hwjUWguLV3JaoYPPJ667J', 
                 'fkGxqf3bEQi4vMZuhvcaCVQMiGoBmdD7fZ3iUMx3hiPD6NMrn3',
                 '930674351585800192-9qA9hKbM4OrBrRtwqtJlsOyzz7eFKKP',
                 'qWJYA0BlLjn4OS69lsi6sDMjD10WxS8Mjs5OgNsSt11Sx')

pager = TwitterPager(api, 'search/tweets', {'q':SEARCH_TERM, 'count':100, 'result_type':'popular'})

result = []
for item in pager.get_iterator(wait=5.5):
    result.append(item)

print(len(result))

SEARCH_TERM = '#Starbucks lang:en'
pager = TwitterPager(api, 'search/tweets', {'q':SEARCH_TERM, 'count':100, 'result_type':'popular'})
for item in pager.get_iterator(wait=5.5):
    result.append(item)
print(len(result))

SEARCH_TERM = 'Starbucks lang:en'
pager = TwitterPager(api, 'search/tweets', {'q':SEARCH_TERM, 'count':100, 'result_type':'popular'})
for item in pager.get_iterator(wait=5.5):
    result.append(item)
print(len(result))
```

```python
# mixed ones

from TwitterAPI import TwitterAPI, TwitterPager

SEARCH_TERM = '#Starbucks lang:en'

api = TwitterAPI('3Ya8hwjUWguLV3JaoYPPJ667J', 
                 'fkGxqf3bEQi4vMZuhvcaCVQMiGoBmdD7fZ3iUMx3hiPD6NMrn3',
                 '930674351585800192-9qA9hKbM4OrBrRtwqtJlsOyzz7eFKKP',
                 'qWJYA0BlLjn4OS69lsi6sDMjD10WxS8Mjs5OgNsSt11Sx')

pager = TwitterPager(api, 'search/tweets', {'q':SEARCH_TERM, 'count':100, 'result_type':'mixed'})

count = 0
result = []
for item in pager.get_iterator(wait=5.5): # 180 requests per 15 minutes
    result.append(item)
    count = count + 1
    if (count % 100 == 0):
        print(result[-1]['created_at'])

print(len(result))
```

```python
import json
with open('starbucks_mixed.json', 'w') as f:
    json.dump(result, f)
```
