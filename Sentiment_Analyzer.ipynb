## sentiment analyzer VADER
(https://github.com/cjhutto/vaderSentiment)
(http://datameetsmedia.com/vader-sentiment-analysis-explained/)

```python
# remove 'RT' from tweet text
# might do sampling within each 10 minutes period

import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

def score(text):
    try: 
        ss = sid.polarity_scores(text)
        return ss['compound']
    except: 
        return 0

def remove_RT(text):
    index = text.find('RT')
    if index != -1:
        text = text[index+3:]
    return text
```

```python
df_timeline = pd.read_csv('tesla_timeline_text.csv', dtype={'created_at':str, 'id_str':str})
df_timeline.fillna('', inplace=True)
df_timeline['text'] = df_timeline['text'].apply(remove_RT)
df_timeline['compound'] = df_timeline['text'].apply(score)
df_timeline['label'] = 'timeline'
df_timeline.head()
```

```python
df_popular = pd.read_csv('tesla_popular_text.csv', dtype={'created_at':str, 'id_str':str})
df_popular.fillna('', inplace=True)
df_popular['text'] = df_popular['text'].apply(remove_RT)
df_popular['compound'] = df_popular['text'].apply(score)
df_popular['label'] = 'popular'
df_popular.head()
```

```python
df_mixed = pd.read_csv('tesla_mixed_text.csv', dtype={'created_at':str, 'id_str':str})
df_mixed.fillna('', inplace=True)
df_mixed['text'] = df_mixed['text'].apply(remove_RT)
df_mixed['compound'] = df_mixed['text'].apply(score)
df_mixed['label'] = 'mixed'
df_mixed.head()
```

```python
df = pd.concat([df_timeline, df_popular, df_mixed])
df = df.drop_duplicates(subset=['id_str'])
df = df.sort_values(by=['created_at'], ascending=False)
df
```

```python
df.to_csv('tesla.csv', index=False)
```
